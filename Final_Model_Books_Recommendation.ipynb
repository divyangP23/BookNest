```python
book_ratings = new_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
book_ratings.head()
```

```python
book_avg_ratings = new_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
book_avg_ratings.rename(columns={'Book-Rating':'Avg-Rating'},inplace=True)
book_avg_ratings.head()
```

```python
popular_ratings = book_ratings.merge(book_avg_ratings,on="Book-Title")
popular_ratings.head()
```

```python
final_popular_books = popular_ratings[(popular_ratings['Avg-Rating']>=3.7) & (popular_ratings['Book-Rating']>=250)]
final_popular_books = final_popular_books.sort_values('Avg-Rating', ascending=False)
final_popular_books.head(50)
```

```python
final_popular_books.duplicated().sum()
```

```python
final_popular_books.isnull().sum()
```

```python
final_popular_books = final_popular_books.merge(books,on="Book-Title").drop_duplicates('Book-Title')
final_popular_books.head()
```

```python
final_popular_books = final_popular_books[['Book-Title','Avg-Rating','Book-Rating','Book-Author','Year-Of-Publication','Publisher','Image-URL-L']]
```

## COLLABARTIVE BASED MODEL

```python
# our approach is to use only those users who have voted min on 200 books and that books on which min 50 users hav voted
```

```python
new_ratings.head()
```

```python
x = new_ratings.groupby('User-ID').count()['Book-Rating']>200
educated_users = x[x].index
filtered_ratings = new_ratings[new_ratings['User-ID'].isin(educated_users)]
filtered_ratings.head()
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
rated_books = y[y].index
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(rated_books)]
final_ratings.head()
```

```python
final_ratings.duplicated().sum()
```

```python
final_table=final_ratings.pivot_table(index='Book-Title',columns = 'User-ID',values='Book-Rating')
```

```python
final_table.fillna(0,inplace=True)
final_table
```

```python
new_final = final_ratings.drop_duplicates(subset='Book-Title').copy()
new_final.head()

```

```python
new_final.to_csv('collab_based_table.csv', index=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(final_table)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index=np.where(final_table.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_books:
        item=[]
        temp_df = books[books['Book-Title']==final_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
final_recommend=recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
final_recommend
```

```python
import pickle
pickle.dump(final_table,open('final_table.pkl','wb'))
pickle.dump(books,open('books.pkl','wb'))
pickle.dump(similarity_score,open('similarity_score.pkl','wb'))
```
